```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
from scipy import constants
```

```python
#Obtengo datos para el efecto 
with open('200T.csv', mode='r') as archivo:
    ArchivoAbierto = csv.reader(archivo) #Usamos la biblioteca csv
    next(ArchivoAbierto,None)            #Me salto el primer renglón
    En, T = [], []
    for renglon in ArchivoAbierto:
        print(renglon[0])
        print(renglon[1])
        #print(type(float(renglon[0])) == float)
        if renglon[1] == '':
            next(ArchivoAbierto,None)
        else:
            En.append(float(renglon[0]))
            T.append(float(renglon[1]))
    En = np.array(En)
    T = np.array(T)
```

```python
d = 0.03 #m
v = d/T

plt.plot(En,v,'.',label='Mediciones')
plt.title('Gráfica a 200 Torr')
plt.ylabel(r'$v [m/s]$'), plt.xlabel(r'$E/n [Td]$')
plt.legend()
plt.grid()
plt.show()
```
