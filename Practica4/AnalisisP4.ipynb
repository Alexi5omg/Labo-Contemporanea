```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
from scipy import constants
```

```python
#Obtengo datos de la polarizacion directa
with open('2v.csv', mode='r') as archivo:
    ArchivoAbierto = csv.reader(archivo) #Usamos la biblioteca csv
    next(ArchivoAbierto,None)            #Me salto el primer renglón
    V2c, V2a, V2b = [], [], []
    for renglon in ArchivoAbierto:
        V2c.append(float(renglon[0]))
        V2a.append(float(renglon[1]))
        V2b.append(float(renglon[2]))

with open('3v.csv', mode='r') as archivo:
    ArchivoAbierto = csv.reader(archivo) #Usamos la biblioteca csv
    next(ArchivoAbierto,None)            #Me salto el primer renglón
    V3c, V3a, V3b = [], [], []
    for renglon in ArchivoAbierto:
        V3c.append(float(renglon[0]))
        V3a.append(float(renglon[1]))
        V3b.append(float(renglon[2]))

with open('4v.csv', mode='r') as archivo:
    ArchivoAbierto = csv.reader(archivo) #Usamos la biblioteca csv
    next(ArchivoAbierto,None)            #Me salto el primer renglón
    V4c, V4a, V4b = [], [], []
    for renglon in ArchivoAbierto:
        V4c.append(float(renglon[0]))
        V4a.append(float(renglon[1]))
        V4b.append(float(renglon[2]))

V2a = np.array(V2a)*-1
V2b = np.array(V2b)*-(1/10)
V3a = np.array(V3a)*-1
V3b = np.array(V3b)*-(1/10)
V4a = np.array(V4a)*-1
V4b = np.array(V4b)*-(1/10)
```

```python
plt.plot(V2c,V2a,'.',label='Va')
plt.plot(V2c,V2b,'*',label='Vb')
plt.title('Gráfica a Vf = 2v')
plt.xlabel('V [Volts]'), plt.ylabel(' Vc [Volts]')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(V3c,V3a,'.',label='Va')
plt.plot(V3c,V3b,'*',label='Vb')
plt.title('Gráfica a Vf = 3v')
plt.xlabel('V [Volts]'), plt.ylabel(' Vc [Volts]')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(V4c,V4a,'.',label='Va')
plt.plot(V4c,V4b,'*',label='Vb')
plt.title('Gráfica a Vf = 4v')
plt.xlabel('V [Volts]'), plt.ylabel(' Vc [Volts]')
plt.legend()
plt.grid()
plt.show()
```
