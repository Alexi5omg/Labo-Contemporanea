```python
#Calculo de las velocidades
d = 0.03 #m
vlist = []
for i in range(len(Tlist)):
    vlist.append(d/Tlist[i])
#Para las graficas
formatlist = ['bx-','g*-','rs-','c+-','md-','yp-','kv-','^-']
Presionlist = ['200.2 Torr','182.0 Torr','162.4 Torr','142.6 Torr','101.7 Torr','60.8 Torr','31.4 Torr','10.1 Torr']
```

```python
#Graficas por separado

def coord(line,dimension):
  contador =  0
  y = line%(dimension)
  while line >= dimension:
    line = line - (dimension)
    contador += 1
  x = contador
  new_coord = [x,y]
  return new_coord


fig , ejes = plt.subplots(3,3,figsize = (11,7), layout="constrained")
for i in range(len(Tlist)):
    x,y = coord(i,3)
    ejes[x,y].plot(Enlist[i],vlist[i],formatlist[i],label='Mediciones a '+ Presionlist[i])
    ejes[x,y].set_ylabel(r'$v [m/s]$'), ejes[x,y].set_xlabel(r'$E/n [Td]$')
    ejes[x,y].legend(), ejes[x,y].grid()

ejes[2,2].remove()
plt.show()
```

```python
#Para las graficas juntas
for i in range(len(Tlist)):
    plt.plot(Enlist[i],vlist[i],formatlist[i],label='Mediciones a '+ Presionlist[i])

plt.title('Incluyendo todos los datos')
plt.ylabel(r'$v [m/s]$'), plt.xlabel(r'$E/n [Td]$')
plt.legend()
plt.grid()
plt.show()
```
