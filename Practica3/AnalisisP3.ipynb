```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from scipy import constants
```

## V_diodo vs Nu

```python
#Obtengo datos del V_diodo y de la Frecuencia
with open('Vdiodo-vs-Frec.csv', mode='r') as archivo:
    ArchivoAbierto = csv.reader(archivo) #Usamos la biblioteca csv
    next(ArchivoAbierto,None)            #Me salto el primer renglón
    Vd, inc_Vd, Frec = [], [], []
    for renglon in ArchivoAbierto:
        Vd.append(float(renglon[3]))
        inc_Vd.append(float(renglon[4]))
        Frec.append(float(renglon[5]))
```

```python
plt.errorbar(Vd,Frec,xerr=inc_Vd,fmt='.',label=r'Mediciones')
plt.title(r'$\nu$ vs. $V_\text{diodo}$')
plt.xlabel(r'$V_\text{diodo}$ [Volt]'), plt.ylabel(r'$\nu$ [$1/s$]')
plt.legend()
plt.grid()
plt.show()
```

```python
#Regresión lineal usando sklearn

Vd_reshape = np.array(Vd).reshape(-1,1) #Reescalamos V_d porque en X se necesita un arreglo 2D

model = LinearRegression() #Invocamos al modelo
model.fit(Vd_reshape, Frec) #Entrenamos al modelo

Frec_reg = model.predict(Vd_reshape) #Hacemos las predicciones (la regresión lineal para la Frecuencia como función de V_d)
```

```python
plt.errorbar(Vd,Frec,xerr=inc_Vd,fmt='.',label=r'Mediciones')
#plt.plot(Vd,Frec,'.',label=r'Mediciones')
plt.plot(Vd,Frec_reg,label='Ajuste lineal')
plt.title(r'$\nu$ vs. $V_\text{diodo}$')
plt.xlabel(r'$V_\text{diodo}$ [Volt]'), plt.ylabel(r'$\nu$ [$1/s$]')
plt.legend()
plt.grid()
plt.show()
```

```python
m = model.coef_[0]     #m: pendiente (e/h)
#b = model.intercept_   #b: ordenada al origen en A
razon_experimental = 1/m

#Calculo exacto de razon
e = constants.e
h = constants.h
razon_real = h / e

print('La razón REAL entre las constantes es de:', razon_real, 'en unidades del SI')
print('La razón EXPERIMENTAL entre las constantes es:',razon_experimental, 'en unidades del SI')
print('Tenemos un error relativo:',(razon_real-razon_experimental)/razon_experimental)
```
